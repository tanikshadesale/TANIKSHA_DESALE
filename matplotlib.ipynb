```python
import matplotlib.pyplot as plt
values=[5,6,3,7,2]
names=["A","B","C","D","E"]
plt.bar(names,values,color="violet")
plt.show()
```

```python
import matplotlib.pyplot as plt


bars= plt.bar([5,6,3,7,2],["A","B","C","D","E"])
bars[0].set_color("violet")
bars[1].set_color("blue")
bars[2].set_color("grey")
bars[3].set_color("red")
bars[4].set_color("green")
plt.show()
```

```python
import matplotlib.pyplot as plt
age=(10,23,14,21,40,80,96,34)
range=(0,100)
bins=10
plt.hist(age,bins,range,color="green",histtype="bar",edgecolor="black")
plt.xlabel('age')
plt.ylabel('no. of people')
plt.title('my histrogram')
plt.show()
```

```python
import matplotlib.pyplot as plt
x_values=[0,1,2,3,4,5]
y_values=[0,1,4,9,12,23]
plt.scatter(x_values,y_values,label="stars",s=30,color="blue",marker="*")
plt.show()
```

```python
import matplotlib.pyplot as plt
activities=['eat','sleep','work','play']
slices=[3,8,3,4]
colors=["pink",'violet','LIME','magenta']
plt.pie(slices,labels=activities, colors=colors, startangle=90, shadow=True, explode=(0,0,0.1,0),radius=1.3,autopct='%1.lf%%')
plt.show()
```
