```python
import pandas as pd

# first data frame
df1 = pd.DataFrame({"fruit" : ["apple", "banana", "mango"],"market_price" : [21, 14, 35]})
display("first DataFrame")
display(df1)

# second Data Frame
df2 = pd.DataFrame({"fruit" : ["banana", "apple", "mango"],"wholesaler_price" : [67, 70, 71]})
display("second DataFrame")
display(df2)

# joining Data Frames
display("merged DataFrame")
pd.merge(df1, df2, on = "fruit", how = "inner")
```
